```python
# -*- coding:utf-8 -*-

import urllib
import urllib2

url_ip='http://httpbin.org/ip'
url_get='http://httpbin.org/get'

def usurlib2():
    response=urllib2.urlopen(url_ip)
    print('response headers:')
    print(response.info())
    print('response body:')
    print(''.join([line for line in response.read()]))

def url_params_urllib2():
    #构造请求参数
    params=urllib.urlencode({'params1':'hello','params2':'world'})
    print(params)
    #发送请求
    response=urllib2.urlopen('?'.join([url_get,'%s'])% params)
    print('response headers:')
    print(response.info())
    print('response body:')
    print(''.join([line for line in response.read()]))
    
    
    
if __name__=='__main__'   :
    print(usurlib2()) 
    print('a')
    print(url_params_urllib2())
```

```python
# -*- coding:utf-8 -*-

import requests

url_ip='http://httpbin.org/ip'
url_get='http://httpbin.org/get?'

def usurlib2():
    response=requests.get(url_ip)
    print('response headers:')
    print(response.headers)
    print('response body:')
    print(response.text)

def url_params_urllib2():
    #构造请求参数
    params={'params1':'hello','params2':'world'}
   
    #发送请求
    response=requests.get(url_get,params=params)
    print('response headers:')
    print(response.headers)
    print(response.status_code)
    print('response body:')
    print(response.text)

    
    
if __name__=='__main__'   :
    print(usurlib2()) 
    print('a')
    print(url_params_urllib2())
```
